```python
import pandas as pd
```

```python
crossings = pd.read_csv("./data/Monthly_Table_Full_Data_data.csv")
crossings = crossings[["Port Name", "Measure", "Month of Date", "Port", "Port Code", "Value"]]
crossings.groupby(by = ["Port Code", "Month of Date"])["Value"].sum()
crossings = pd.DataFrame(crossings.groupby(by = ["Port Code"])["Value"].sum()).reset_index().rename(columns = {"Port Code":"port_code", "Value":"num_mig"})
crossings.head()
```

```python
coords = pd.read_csv("./data/Canada_and_Mexico_Border_Crossings.csv")
coords = coords[["X", "Y", "portcode"]]
coords = coords.rename(columns = {'portcode': "port_code"})
coords = coords.drop_duplicates(subset = "port_code")
coords.head()
```

```python
comb = pd.merge(crossings, coords, on = "port_code")
comb.head()
```

```python
comb.to_csv("./data/mexico_border_2010.csv", index = False)
```
